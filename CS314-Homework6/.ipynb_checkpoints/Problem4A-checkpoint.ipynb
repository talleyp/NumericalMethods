```julia
function euler(t,h)
    y = 3/2 * t^(1/3)
    return y + h*y
end
```

```julia
h=0.1
ts = 0:h:3
n = length(ts)
y = zeros(n,1)
actual = zeros(n,1)
@printf("Time \t Approx \t Exact \t\t Error Percent \n")
@printf("%d \t %.8f \t %.8f \t %.4f\n", ts[1], y[1], actual[1], abs(actual[1]-y[1]))
for i=2:n
    y[i] = y[i-1] + euler(y[i-1],h)
    actual[i]=ts[i]^(3/2)
    @printf("%.2f \t %.8f \t %.8f \t %.4f\n", ts[i], y[i], actual[i], abs(actual[i]-y[i])/actual[i]*100)
end
    
```

```julia
using Plots
plot(actual,ts,label="actual")
plot!(y,ts,label="Euler")
```
