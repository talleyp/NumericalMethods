```julia
using Plots, ODE;
```

```julia
function love(t,f)
    (x, y) = f
    dx_dt = -0.2*y
    dy_dt = 0.8*x -0.1*y
    
    [dx_dt; dy_dt]
end;
```

```julia
start = [2.0; 0.0]
ts = [0.0; 60.0];
```

```julia
t, res = ode45(love, start, ts)

r = map(x -> x[1], res)
j = map(x -> x[2], res);
```

```julia
plot(t, r, label="Romeo")
plot!(t, j, label="Juliet")
xlabel!("t")
```
