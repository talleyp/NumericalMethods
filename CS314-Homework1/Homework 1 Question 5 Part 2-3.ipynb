```julia
using Plots

x = linspace(0, 2*pi, 1000);

y1 = sin(1 * x);
y2 = sin(2 * x);
y3 = sin(3 * x);
y4 = sin(4 * x);
y5 = sin(5 * x);

plot(x, y1)
plot!(x, y2)
plot!(x, y3)
plot!(x, y4)
plot!(x, y5)
```

```julia
using Plots

plotly(size=(500,500)) # use plotly for zooming and 500-by-500 for equal size
theta = linspace(0, 2*pi, 10000)
r1 = sqrt(2)
x1 = 2. + r1*cos(theta)
y1 = 1. + r1*sin(theta)

r2 = sqrt(3.5)
x2 = 2.5 + r2*cos(theta)
y2 = r2*sin(theta)
plot(x1,y1)
plot!(x2,y2)
```
